# Removing Outliers

```python
#outliers are those values whose z-score is less than -3 and greater than 3
from scipy import stats
z=np.abs(stats.zscore(train_data))
z
```

```python
# Definea threshold for identifying outlier
threshold=3
np.where(z>threshold)
```

```python
# Get clean data by removing outliers
df = train_data[(z<=3).all(axis=1)]
df
```
